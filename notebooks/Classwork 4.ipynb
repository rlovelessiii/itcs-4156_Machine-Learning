```python
import sklearn.datasets as data
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
wine = data.load_wine()
```

```python
# Now let us convert the Data Set
df = pd.DataFrame(wine.data, columns = wine.feature_names)
```

```python
df['wine_type'] = wine.target
```

```python
df.head()
```

```python
# Let us try to predict the Alcohol Values using Single Value
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
# The variable "reg" is an instentiator 
reg
```

```python
x = df['proline']
y = df['alcohol']
```

```python
x
```

```python
y
```

```python
# Since we will be using values of "x" we need it in 2D Array
x = df[['proline']]
```

```python
x
```

```python
#Fit linear model
reg.fit(x, y)
```

```python
#Predict using the linear model
pred = reg.predict(x)
```

```python
pred
```

```python
from sklearn import metrics
#module includes score functions, performance metrics 
#and pairwise metrics and distance computations.
```

```python
#Calculate R-Squared Value
metrics.r2_score(y, pred)
```

```python
# Since we are predicting Alcohol Values, we need to drop this feature and also Wine_Type
x = df.drop(['alcohol', 'wine_type'], axis =1)
```

```python
x
```

```python
reg.fit(x, y)
```

```python
pred = reg.predict(x)
```

```python
pred
```

```python
metrics.r2_score(y, pred)
```

```python
# Let us learn about our Classification by using Logistic Regression
# In Logistic Regression the values remain between 0 and 1
# Binary dependent variable, can be extended to predict outputs with more than 2 variables
# This is also called GLM (Generalized Linear Models)
```

```python
# Let us Predict Wine_Type
```

```python
x = df.drop('wine_type', axis=1)
y = df ['wine_type']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf
```

```python
clf.fit(x,y)
```

```python
pred = clf.predict(x)
```

```python
pred
```

```python
metrics.accuracy_score(y, pred)
```

```python
# This is very near to 1, and looks excellent as it is 97%
```
